```python
# E-Commerce Shopping Cart

products = {
    1: {"name": "T-Shirt", "price": 499},
    2: {"name": "Shoes", "price": 899},
    3: {"name": "Backpack", "price": 699}
}

cart = []

def show_products():
    print("\n Available Products:")
    for pid, info in products.items():
        print(f"{pid}. {info['name']} - ₹{info['price']}")

def add_to_cart():
    pid = int(input("\nEnter product ID to add to cart: "))
    if pid in products:
        cart.append(products[pid])
        print(f"{products[pid]['name']} added to cart.")
    else:
        print("Invalid product ID.")

def view_cart():
    print("\nYour Cart:")
    if not cart:
        print("Cart is empty.")
        return

    total = 0
    for item in cart:
        print(f"{item['name']} - ₹{item['price']}")
        total += item['price']
    print(f"Total Amount: ₹{total}")
    return total

def checkout():
    total = view_cart()
    if total:
        choice = input("\nProceed to payment? (yes/no): ")
        if choice.lower() == "yes":
            print("Processing payment...")
            print("Payment Successful! Thank you for shopping.")
        else:
            print("Payment Cancelled.")
    else:
        print("Nothing to checkout.")
```

```python
# Main program loop
while True:
    print("\n--- E-Commerce Menu ---")
    print("1. Show Products")
    print("2. Add to Cart")
    print("3. View Cart")
    print("4. Checkout")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        show_products()
    elif choice == "2":
        show_products()
        add_to_cart()
    elif choice == "3":
        view_cart()
    elif choice == "4":
        checkout()
    elif choice == "5":
        print("Exiting... Visit again!")
        break
    else:
        print("Invalid choice. Please try again.")
```
